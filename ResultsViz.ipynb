```python
import numpy as bp
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("errors.csv")
df.head()
```

```python
#MAE Visualisation
#Create plot
df=df.sort_values('MAE')
plt.figure(figsize=(8,6))
plt.scatter(df['MAE'],df['Model'],color='red')
plt.grid(True)

# Add labels and title
plt.xlabel('MAE')
plt.ylabel('Model')
plt.title('MAE Values by Model')

#show the plot
plt.show()
```

```python
#RMSE Visualisation
#Create plot
df=df.sort_values('RMSE')
plt.figure(figsize=(8,6))
plt.scatter(df['RMSE'],df['Model'],color='red')
plt.grid(True)

# Add labels and title
plt.xlabel('RMSE')
plt.ylabel('Model')
plt.title('RMSE Values by Model')

#show the plot
plt.show()
```
